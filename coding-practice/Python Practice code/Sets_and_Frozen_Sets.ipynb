```python
# 1.create any set anf try to use frozenset(setname)

# 2.Find the elements in a given set that are not in another set
#     set1 = {1,2,3,4,5}
#     set2 = {4,5,6,7,8}

#     diffrence between set1 and set2 is {1,2,3}
set1 = {1,2,3,4,5}
set2 = {4,5,6,7,8}
print(set1)
print(set2)
print("Differnce of set1 and set2 using Difference()")
print(set1.difference(set2))
print("Differnce of set1 and set2 using Difference()")
print(set2.difference(set1))
print("Differnce of set1 and set2 using - Operater()")
print(set1 - set2)
print("Differnce of set1 and set2 using - Operater()")
print(set2 - set1)
```
