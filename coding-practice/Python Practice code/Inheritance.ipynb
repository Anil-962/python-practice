```python
# 1.create inheritance using animal Dog relation.
# for example,
#     Animal and Dog both has same habitat so create a method for habitat
# 2.use super() constructor for calling parent constructor.
# class Animal:
#     #code

# class Dog(Animal):
#     super()-it refers Animal class,now you can call Animal's methods.

class Animal:
  def __init__(self,name,habitat):
    self.name = name
    self.habitat = habitat
  def print_all(self):
    print(f'Animal name is {self.name} the habitat is {self.habitat}')
  def sound():
    print("Animal Sound")

class Dog(Animal):
  def __init__(self):
    super().__init__("Dog","Land")
  def sound(self):
    print("Woof")
x = Dog()
x.print_all()
x.sound()

```
