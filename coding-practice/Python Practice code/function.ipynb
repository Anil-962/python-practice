```python
# Write a function called calculate_area that takes base and height as
# an input and returns and area of a triangle. Equation of an area of a triangle is,

def calculate_area(base,height):
  return (1/2)*base*height
base = int(input("Enter the base "))
height = int(input("Enter the height "))
area = calculate_area(base,height)
print(area)
```

```python
# Modify above function to take third parameter shape type. It can be either "triangle" or "rectangle".
#  Based on shape type it will calculate area. Equation of rectangle's area is,
def calculate_area(base,height,shape="Triangle"):

  if shape=="Triangle":
     return (1/2)*base*height
  elif shape=="rectangle":
     return base*height
  else:
    print("Invalid input")
base = int(input("Enter the base "))
height = int(input("Enter the height "))
shape = input()
area = calculate_area(base,height,shape)
print(area)

```

```python
#Write a function called print_pattern that takes integer number
#as an argument and prints following pattern if input number is 3,
def pattern(n=5):
  for i in range(n):
    s =''
    for j in range(i+1):
      s=s+"*"
    print(s)
pattern(3)
```
